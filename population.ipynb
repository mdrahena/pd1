```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"C:\Users\mdrah\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_2457130\API_SP.POP.TOTL_DS2_en_csv_v2_2457130.csv"

data = pd.read_csv(file_path, skiprows=4)

# Check the first few rows and columns of the dataframe to understand its structure
print(data.head())
print(data.columns)

# Assuming the dataset has a column 'Year' and 'Population' for this example
# Modify the column names based on the actual dataset
# For example, let's check if there are columns 'Country Name', '1990', '2000', etc.
# We'll use one country's data as an example (e.g., 'India')

country_data = data[data['Country Name'] == 'India'].T
country_data = country_data[4:]  # Skip the first four rows which are metadata
country_data.columns = ['Population']
country_data.index.name = 'Year'
country_data.reset_index(inplace=True)

years = country_data['Year']
population = country_data['Population'].astype(float)  # Convert to float for plotting

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Distribution Over Years (India)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()

# Create a histogram
plt.figure(figsize=(12, 6))
plt.hist(population, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Histogram of Population Distribution (India)')
plt.show()
```
